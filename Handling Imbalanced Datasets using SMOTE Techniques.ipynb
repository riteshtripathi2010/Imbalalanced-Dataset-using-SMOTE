```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/master/car_evaluation.csv')
data.head()

#model always tends to be more biased towards the majority class

#How to handle that:
    #1. Undersampling
    #out of 1000 data, there are 900 non cancer patients
        #100 are cancer patients
    #we keep, 100 cancers as 100
    #and randomly sample 100 from 900 
    #hence we have 200 records
    #we have 100 non cancer data
    #and 100 cancer data
    #but there is an issue, from 1000 data points, we are loosing 800 datapoints
    #this will decrease on your accuracy
    
    
    #2. Oversampling
        #we keep 900 as 900
        #and we make 100 as 900
            #how do we do that: simple duplication
            #out of 100, we keep removing 30 with replacing
                #we take 30 30 30 randomly and do it for 30 times
                #its random duplication
                
        #from 1000, it becomes data points as 1800
        #this balances the dataset
        #over sampling is the most important techniques
        
    #3. SMOTE
        #in this technique, we generate data
```

```python
data.shape
```

```python
data.outcome.value_counts()
```

```python
X = data.iloc[:,:-1]
y = data.outcome
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
X.loc[:,['buying','maint','lug_boot','safety']] = \
X.loc[:,['buying','maint','lug_boot','safety']].apply(enc.fit_transform)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = \
train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
pd.crosstab(y_test,y_predict)

#we see that right diagnoly are rightly classified
#366 of records are classified as unaccepted 
#10 classified as good and so on
#and all the rest are not classified correctly

#the top row are called predictions
#left colns are called actuals
```

```python
X_train_smote, y_train_smote = smote.fit_resample(X_train.astype('float'),y_train) 
```

```python
from collections import Counter
print("Before SMOTE :" , Counter(y_train))
print("After SMOTE :" , Counter(y_train_smote))

#before smoting, in training data set you had 839 data in unaccepted class 
#accepted had 282 and so on

#and after smoting, all classes have now same number of records
#how this has happend: by creating extrapolating individual classes
#accepted had 282 numbers of data, we use these 282, take avges, interpolate (by taking avg of neighbouring data and create new data till it becomes 839)
```

```python
model.fit(X_train_smote,y_train_smote)
y_predict = model.predict(X_test)
print(accuracy_score(y_test,y_predict))
pd.crosstab(y_test,y_predict)
#we see now the accuracy has gone up as compared with previous
```
